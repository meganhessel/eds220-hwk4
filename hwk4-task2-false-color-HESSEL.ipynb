## 5. True Color Image

GOAL: Create a new true color image that gets plotted without warnings.

```python
# Create an RGB image with selected bands... oh no! There are NAs, extreme values, and warnings!

(landsat[['red', 'green', 'blue']] # Select the red, green, and blue variables 
    .to_array() # Convert to array 
    .plot.imshow()) # Create an RGB image
```

```python
# Identifying what bands have nans 
(landsat[['red', 'green', 'blue']]
    .to_array() 
    .isnull().sum()) # Gathering all values where isnull = TRUE 

# Array(110) have a nan! 
```

```python
# Substitute the any nan values for 0 AND Create a new true color image with NO warnings 
(landsat[['red', 'green', 'blue']]
    .to_array() 
    .fillna(0) # nan values to 0 
    .plot.imshow(robust = True)) # Create an RGB image! robust = True elimates all outlier / extereme values 
    # .plot.imshow(vmin=7000, vmax=15500))
```

### Comparing between plots 

The first plot contains extreme/outlier values from the clouds' RGD measurements that mask and distort the real raster image.

"Robust = True" or setting informed vmax and vmins elimiates the extreme values, making the map show the proper raster of the land (illustrated in the second figure). Specifically, "Robust = True" takes the values between the 2nd and 98th percentile, excluding the outliers. 

## 6. False color image

GOAL: Without creating any new variables, create a false color image by plotting the short-wave infrared (swir22), near-infrared, and red variables (in that order).

```python
# Plotting the short-wave infrared (swir22), near-infrared, and red variables 

(landsat[['swir22', 'nir08', 'red']]
    .to_array()  
    .fillna(0)
    .plot.imshow(robust = True)) # Create an RGB image

# red band = anything reflected swir22 
```

## 7. Map 

1. Create a map showing the shortwave infrared/near-infrared/red false color image together with both fire perimeters. Customize it appropriately including, at least, an informative title and legend. You may also want to include text on the map to identify which fire is which.

2. Write a figure description for the map including a brief explanation of how false color imagery is being used.

```python
# Transform fire border polygons' crs to match landsat's crs   
eaton.to_crs(landsat.rio.crs, inplace = True)
pal.to_crs(landsat.rio.crs, inplace = True)
```

```python
# Create an empty plot 
fig, ax = plt.subplots()

# No axises
ax.axis('off')

# RGB raster 
(landsat[['swir22', 'nir08', 'red']]
    .to_array() 
    .fillna(0)
    .plot.imshow(ax = ax, robust = True)) # nan values to 0

# Fire borders 
eaton.plot(ax = ax, color = 'none', edgecolor = 'darkred')
pal.plot(ax = ax, color = 'none', edgecolor = 'darkblue')

# Add title 
ax.set_title('SWIR/NIR/Red False Color Imagery \n of Palisades and Eaton Wildfires of 2025')

# Add labels 
ax.text(.1, .51, 'Palisades Fire Boundary', 
        transform = ax.transAxes, # controls the position
        fontsize = 9, 
        color = 'white',
        fontweight = 'semibold', 
        bbox = dict(boxstyle = 'round', facecolor = "black", alpha=0.5)
        )

ax.text(.70, .80, 'Eaton Fire Boundary', 
        transform = ax.transAxes, # controls the position
        fontsize = 9, 
        color = 'white', 
        fontweight = 'semibold',
        bbox = dict(boxstyle = 'round', facecolor = "black", alpha=0.5)
        )


# Legend 
pal_legend = mpatches.Patch(facecolor = "darkblue", edgecolor = "black", label = "Palisades Fire Border")
easton_legend = mpatches.Patch(facecolor = "darkred", edgecolor = "black", label = "Eaton Fire Border")

ax.legend(handles = [pal_legend, easton_legend], loc = 'upper left', fontsize = 'x-small')

# Scalebar
ax.add_artist(ScaleBar(1, location = 'lower right', box_color="0.9", border_pad = 0.5, box_alpha = 0.7))

plt.show()
```

### Beief Explaination

Near infrared (NIR) wavelengths reflects healthy vegetation; lower NIR measurements can show unhealthy plant-life conditions, such as drought stress or deforestation. Assigning the NIR to the green channel emphasizes the vegetation health. 

Short wave infrared (SWIR) wavelengths reflects fire and recently burned areas. By assigning SWIR to the red channel, the fire markings can be highlighted, espically in contrast to the NIR (green channel). 

In conclusion, false color imagery is used to emphasises the burn damages from the fires in comparison to the neighboring healthy vegetation.
